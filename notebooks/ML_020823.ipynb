Загружается предобработанный датасет

```python
kc_house_data = pd.read_csv('../data/kc_house_data_preprocessed.csv').sample(10000)
kc_house_data.drop(["Unnamed: 0"], axis=1, inplace=True)
kc_house_data
```

Сперва выделяется целевой признак и выносится за датасет, в котором остаются предикаторы. Целевым признаком будет столбец <em>price</em>.

### Линейная регрессия

С помощью библиотеки sklearn происходит обучение модели и получение предсказаний для тестовой выборки.

```python
lr = LinearRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)
lr.coef_
```

Вычисление и вывод качественных метрик.

```python
print("MAE:", MAE(y_test, y_pred))
print("MSE:", MSE(y_test, y_pred))
print("RMSE:", RMSE(y_test, y_pred))
print("MAPE:", MAPE(y_test, y_pred))
print("R^2:", R2(y_test, y_pred))
```

### Регуляризация (Линейная регрессия)

```python
params = {'alpha': np.arange(0.1, 1, 0.1)}
```

```python
ridge = GridSearchCV(Ridge(), params).fit(X_train, y_train)
y_pred = ridge.predict(X_test)
print("MAE:", MAE(y_test, y_pred))
print("MSE:", MSE(y_test, y_pred))
print("RMSE:", RMSE(y_test, y_pred))
print("MAPE:", MAPE(y_test, y_pred))
print("R^2:", R2(y_test, y_pred))
ridge.best_params_
```

```python
lasso = GridSearchCV(Lasso(tol=0.1), params).fit(X_train, y_train)
y_pred = lasso.predict(X_test)
print("MAE:", MAE(y_test, y_pred))
print("MSE:", MSE(y_test, y_pred))
print("RMSE:", RMSE(y_test, y_pred))
print("MAPE:", MAPE(y_test, y_pred))
print("R^2:", R2(y_test, y_pred))
lasso.best_params_
```

### Написание модели

Функцию потерь $L$ можно переписать в виде $L(w) = \frac{1}{m} \sum_{i=1}^{m} ({y_{i} - x_{i}w})^2$, где $w$ - вектор. Беря частные производные по переменным $w_{j}$ получим градиент:
$$\frac{\partial L(w)}{\partial w_{j}} = 
\frac{1}{m}\sum_{i=1}^{m} 2(y_{i} - x_{i}w)\frac{\partial (y_{i} - x_{i}w)}{\partial w_{j}} =
\frac{1}{m}\sum_{i=1}^{m} 2(y_{i} - x_{i}w)(-x_{ij})$$

И, наконец, делая шаг вдоль антиградиента, получаем формулу пересчёта весов:
$$w_{j}^{new} = w_{j}^{old} - \lambda\frac{1}{m}\sum_{i=1}^{m} 2(y_{i} - x_{i}w)(-x_{ij})$$
где $\lambda$ - значение шага

Что можно переписать в векторном виде как:
$$w^{new} = w^{old} - 2\frac{\lambda}{m}(Y - Xw^{old})(-X^{T})$$

В случае регуляризаций L1 и L2 формула пересчётов соответственно будет выглядеть так:<br>
$L1$: &emsp; $w^{new} = w^{old} - (\frac{\lambda}{m}(Xw^{old}-Y)(-X^{T}) + sign({w^{old}}))$<br>
$L2$: &emsp; $w^{new} = w^{old} - (\frac{\lambda}{m}(Xw^{old}-Y)(-X^{T}) + 2{w^{old}})$<br>

Именно эти формулы будут использоваться при обучении моей модели.
